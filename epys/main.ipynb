### Giriş
Kullanıcı adınızla Web Servis için değişken oluşturmanız gerekmektedir. Bunun için EPYS'de kayıtlı olan kullanıcınızla giriş yapmanız gereklidir. Organizasyonunuz ID'sini int formatında girmeniz gerekmektedir.
##### Zorunlu Parametreler
 - username: str
 - organizationId: int

```python
ws = WebService(username = "USERNAME", organizationId = 123456789)
```

Sisteme login fonksiyonuyla giriş yapılması gerekemektedir.
 - Giriş yaptığınızda TGT alınmış olur.
 - TGT'nin süresi 45 Dakikadır, bu süre içerisinde tekrar giriş yapmanıza gerek yoktur. 
 - 45 dakika dolduktan sonra tekrar sisteme giriş yapmanız gerekemektedir.

Sisteme giriş yaparken direkt olarak şifrenizi str formatında fonksiyona girebilirsiniz. Eğer login (giriş) fonksiyonunu şifre girmeden çalıştırdığınızda input olarak şifreniz istenir, herhangi bir yerde kullanılmaz, depolanmaz, paylaşılmaz.

Her iki örnekte paylaşılmıştır.

```python
ws.login("PASSWORD")
```
